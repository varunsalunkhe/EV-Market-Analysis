```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns 
plt.style.use("ggplot") 
import warnings
warnings.filterwarnings("ignore")
data = pd.read_csv("2-wheeler-EV-bikewale.csv")
print(data.shape)
data.head()
```

```python
print(data.columns)
```

```python
data.info()
```

```python
# Droping the columns whose null values are more than 200

lst=data.columns
for i in lst:
    if data[i].isna().sum() > 200:
        data.drop(i, axis=1, inplace=True)
data.head()
```

```python
data.shape
```

```python
# droping review column because didn't contain important information
data.drop(["review"], axis =1, inplace=True)
```

```python
data.head()
```

```python
# sum of NaN values in each columns

lst2= data.columns
for i in lst2:
    print(i, data[i].isna().sum() )
```

```python
data.drop(0,axis=0,inplace=True)
data.head()
```

```python
# Handling Ridden for column

data["Ridden for"].value_counts()
```

```python
ind=data[(data["Ridden for"] == "Never owned")].index
data.drop(ind,axis=0,inplace=True)
```

```python
data["Ridden for"].fillna("< 5000 kms", inplace= True)
Riddenfor=data["Ridden for"].value_counts()

# plt.figure(figsize=(15,5))

# plt.subplot(121)
sns.barplot(Riddenfor.index, Riddenfor);

# plt.subplot(122)
# plt.pie(Riddenfor, labels=Riddenfor.index, autopct='%1.1f%%', shadow = True,explode = [0, 0.1,0,0])
```

```python
# Handing Visual Appeal column

data["Visual Appeal"].value_counts()
```

```python
# filling the null values with mean of visual appeal
data["Visual Appeal"].fillna(round(data["Visual Appeal"].mean(),0), inplace= True)
VisualAppeal = data["Visual Appeal"].value_counts()
sns.barplot(VisualAppeal.index, VisualAppeal);
```

```python
# handling Reliability column

data["Reliability"].value_counts()

```

```python
# filling the null values with mean of visual appeal
data["Reliability"].fillna(round(data["Reliability"].mean(),0), inplace= True)
```

```python
Reliability = data["Reliability"].value_counts()

sns.barplot(Reliability.index, Reliability);
```

```python
# Handing Service Experience column

data["Service Experience"].value_counts()
```

```python
# filling the null values with mean of visual appeal
data["Service Experience"].fillna(round(data["Service Experience"].mean(),0), inplace= True)
ServiceExperience = data["Service Experience"].value_counts()

sns.barplot(ServiceExperience.index, ServiceExperience);
```

```python
data.info()
```

```python
# Removing the vechicle catogory for analyse, based on whose sum of vechicle category is gretter than 20

list3= data["Model Name"].value_counts()
list4= list3[list3>20]
list4
```

```python
# The customers, how they used vehicles, how long owned and how long ridden for each vechicle company

list5= list4.index
grp = data.groupby("Model Name")

for i in range(len(list5)):

    a=grp.get_group(list5[i])

    plt.figure(figsize=(20,5))

    plt.subplot(131)
    plt.pie(a["Used it for"].value_counts(),labels=a["Used it for"].value_counts().index, autopct='%1.1f%%', shadow = True)
    # plt.title("Used it for")

    plt.subplot(132)
    plt.pie(a["Owned for"].value_counts(),labels=a["Owned for"].value_counts().index, autopct='%1.1f%%', shadow = True)
    # plt.title("Owned for")

    plt.subplot(133)
    plt.pie(a["Ridden for"].value_counts(),labels=a["Ridden for"].value_counts().index, autopct='%1.1f%%', shadow = True)
    # plt.title("Ridden for")


    plt.suptitle(list5[i], fontsize=30);
```

```python
list6= list3[list3<20].index
gk= grp.mean()
gk.drop(list6,axis=0,inplace=True)
gk.shape
```

```python
# Average rating for all electronic vechicle

plt.figure(figsize=(12,5))
plt.bar(gk.index, gk.rating)
# plt.ylim(0,6)
plt.title("Average rating for all Electronic Vechicle", fontsize = 20)
plt.xticks(rotation=70);
```

```python
# Average Visual Appeal for all electronic vechicle

plt.figure(figsize=(12,5))
plt.bar(gk.index, gk["Visual Appeal"])
plt.title("Average Visual Appeal for all Electronic Vechicle", fontsize = 20)
plt.xticks(rotation=70);
```

```python
# Average Reliability for all electronic vechicle

plt.figure(figsize=(12,5))
plt.bar(gk.index, gk["Reliability"])
plt.title("Average Reliability for all Electronic Vechicle", fontsize = 20)

plt.xticks(rotation=70);
```

```python
# Average Service Experience for all electronic vechicle

plt.figure(figsize=(12,5))
plt.bar(gk.index, gk["Service Experience"])
plt.title("Average Service Experience for all Electronic Vechicle", fontsize = 20)

plt.xticks(rotation=70);
```

```python
# by analysing the data mostly liked or best performer Electronic vechicles

gk["Best performer"]=(gk["Visual Appeal"]+gk["Reliability"]+gk["rating"]+gk["Service Experience"])/4
gk.head()
```
